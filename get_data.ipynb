```python
# 导入tushare
import tushare as ts
# 初始化pro接口
pro = ts.pro_api('2876ea85cb005fb5fa17c809a98174f2d5aae8b1f830110a5ead6211')

# 拉取股票列表数据
'''
ts_code	str	TS代码
symbol	str	股票代码
name	str	股票名称
area	str	地域
industry str	所属行业
cnspell	str	拼音缩写
market	str	市场类型
list_date	str	上市日期
delist_date	str	退市日期
act_name	str	实控人名称
act_ent_type	str	实控人企业性质
list_status	str	上市状态 L上市 D退市 P暂停上市
'''
df = pro.stock_basic(**{
    "ts_code": "",
    "name": "",
    "exchange": "",
    "market": "",
    "is_hs": "",
    "list_status": "",
    "limit": "",
    "offset": ""
}, fields=[
    "ts_code",
    "symbol",
    "name",
    "area",
    "industry",
    "cnspell",
    "market",
    "list_date",
    "act_name",
    "act_ent_type",
    "delist_date",
    "list_status"
])
print(df)

        
```

```python
# 获取交易日期，便于循环遍历
'''
exchange	str	Y	交易所 SSE上交所 SZSE深交所
cal_date	str	Y	日历日期
is_open	str	Y	是否交易 0休市 1交易
pretrade_date	str	Y	上一个交易日
'''
cal = pro.trade_cal(exchange='', start_date='20180101', end_date='20250430', fields=["exchange","cal_date","is_open","pretrade_date"], is_open='1')
print(len(cal))
```

```python
import time
import pandas as pd
# 拉取股本/估值数据
'''
ts_code	str	TS股票代码

trade_date	str	交易日期

close	number	当日收盘价

turnover_rate_f	number	换手率(自由流通股)

volume_ratio	number	量比

pe	number	市盈率（总市值/净利润）

pe_ttm	number	市盈率（TTM）

pb	number	市净率（总市值/净资产）

ps	number	市销率

ps_ttm	number	市销率（TTM）

dv_ratio	number	股息率（%）

dv_ttm	number	股息率（TTM） （%）

float_share	number	流通股本

circ_mv	number	流通市值

limit_status	int	涨跌停状态
'''
# 拉取数据
df_final = []
for td in list(cal['cal_date']):
    if int(td) >= 20190425:
        continue
    df = pro.daily_basic(**{
        "ts_code": "",
        "trade_date": "",
        "start_date": td,
        "end_date": td,
        "limit": "",
        "offset": ""
    }, fields=[
        "ts_code",
        "trade_date",
        "close",
        "turnover_rate_f",
        "volume_ratio",
        "pe",
        "pe_ttm",
        "pb",
        "ps",
        "ps_ttm",
        "dv_ratio",
        "dv_ttm",
        "float_share",
        "circ_mv",
        "limit_status"
    ])
    print(td,' finished with ',len(df))
    time.sleep(1)
    df_final.append(df)
     
df_final = pd.concat(df_final, ignore_index=True)
df_final.to_pickle('stock_value.pkl')
print(df_final.head())
```

```python
# 提取行情日线数据
# 导入tushare
import tushare as ts
import pandas as pd
import time
# 初始化pro接口
pro = ts.pro_api('2876ea85cb005fb5fa17c809a98174f2d5aae8b1f830110a5ead6211')
cal = pd.read_pickle('calendar.pkl')
'''
ts_code	str	股票代码

trade_date	str	交易日期

open	float	开盘价

high	float	最高价

low	float	最低价

close	float	收盘价

pre_close	float	昨收价

change	float	涨跌额

pct_chg	float	涨跌幅

vol	float	成交量

amount	float	成交额
'''
df_final = []
for td in list(cal['cal_date']):
    df = pro.daily(**{
        "ts_code": "",
        "trade_date": "",
        "start_date": td,
        "end_date": td,
        "offset": "",
        "limit": ""
    }, fields=[
        "ts_code",
        "trade_date",
        "open",
        "high",
        "low",
        "close",
        "pre_close",
        "change",
        "pct_chg",
        "vol",
        "amount"
    ])
    print(td,' finished with ',len(df))
    time.sleep(0.2)
    df_final.append(df)
     
df_final = pd.concat(df_final, ignore_index=True)
df_final.to_pickle('stock_kline.pkl')
print(df_final.head())
        
```

```python
import tushare as ts
import pandas as pd
# 初始化pro接口
pro = ts.pro_api('2876ea85cb005fb5fa17c809a98174f2d5aae8b1f830110a5ead6211')
#获取申万一级行业列表
df1 = pro.index_classify(level='L1', src='SW2021')

#获取申万二级行业列表
df2 = pro.index_classify(level='L2', src='SW2021')

#获取申万三级级行业列表
df3 = pro.index_classify(level='L3', src='SW2021')
```

```python
# 标注二/三级行业所属一级行业，标注个股所属一级行业
df3['industry_code'] = df3['industry_code'].apply(lambda x: x[:2] + '0' * (len(x) - 2))
industry_dict = df3.set_index('industry_name')['industry_code'].to_dict()
basic = pd.read_pickle('stock_basic.pkl')
basic['industry_l1'] = basic['industry'].map(industry_dict)

df2['industry_code'] = df2['industry_code'].apply(lambda x: x[:2] + '0' * (len(x) - 2))
industry_dict = df2.set_index('industry_name')['industry_code'].to_dict()
basic.loc[basic['industry_l1'].isna(), 'industry_l1'] = basic.loc[basic['industry_l1'].isna(), 'industry'].map(industry_dict)

industry_dict = df1.set_index('industry_name')['industry_code'].to_dict()
basic.loc[basic['industry_l1'].isna(), 'industry_l1'] = basic.loc[basic['industry_l1'].isna(), 'industry'].map(industry_dict)
print(f'缺少{sum(basic['industry_l1'].isna())}条')
basic.head()
```
